# Creación del dataset sintético completo

```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta
```

```python
random.seed(42)
np.random.seed(42)
```

```python
# definimos el tamaño del dataset que queremos crear
n_pacientes = 400
n_citas = 1600
```

```python
# crear las listas base
generos = ["Mujer", "Hombre", "No binario", "Prefiere no decir"]

ocupaciones = [
    "Estudiante",
    "Empleado/a",
    "Autónomo/a",
    "Desempleado/a",
    "Cuidador/a",
    "Jubilado/a",
]

motivos_consulta = [
    "Ansiedad",
    "Estrés académico",
    "Estrés laboral",
    "Depresión leve",
    "Duelo",
    "Autoestima",
    "Problemas de pareja",
    "Orientación familiar",
]

niveles_urgencia = ["Bajo", "Medio", "Alto"]

modalidades = ["Online", "Presencial"]

canales_reserva = ["Teléfono", "WhatsApp", "Web", "Email"]

estados_cita = ["Asistida", "Cancelada", "No asistida", "Reprogramada"]

continuidades = ["Continúa", "Abandona", "Alta terapéutica", "Derivación"]
```

```python
# creamos los IDs de pacientes
ids_pacientes = [f"P{str(i).zfill(4)}" for i in range(1, n_pacientes + 1)]

ids_pacientes[:10]
```

```python
#comprobamos que hay 250 pacientes 
len(ids_pacientes)
```

```python
# creamos datos de pacientes
pacientes = []

for id_paciente in ids_pacientes:
    edad = np.random.randint(18, 76)

    genero = random.choices(generos, weights=[55, 40, 3, 2], k=1)[0]

    if edad <= 25:
        ocupacion = random.choices(ocupaciones, weights=[70, 15, 3, 8, 3, 1], k=1)[0]
    elif edad <= 60:
        ocupacion = random.choices(ocupaciones, weights=[5, 55, 15, 12, 10, 3], k=1)[0]
    else:
        ocupacion = random.choices(ocupaciones, weights=[0, 5, 3, 2, 10, 80], k=1)[0]

    pacientes.append(
        {
            "id_paciente": id_paciente,
            "edad": edad,
            "genero": genero,
            "ocupacion": ocupacion,
        }
    )
```

```python
# lo convertimos en un DataFrame
df_pacientes = pd.DataFrame(pacientes)
df_pacientes.head()
```

```python
df_pacientes. shape
```

```python
# crear función para asigmar mottivo de la consulta
def asignar_motivo(edad, ocupacion):

    if edad <= 25 or ocupacion == "Estudiante":
        motivo = random.choices(
            motivos_consulta, weights=[30, 30, 5, 8, 5, 12, 5, 5], k=1
        )[0]

    elif edad <= 60:
        motivo = random.choices(
            motivos_consulta, weights=[20, 5, 25, 10, 8, 12, 15, 5], k=1
        )[0]

    else:
        motivo = random.choices(
            motivos_consulta, weights=[15, 0, 5, 12, 30, 8, 5, 25], k=1
        )[0]

    return motivo
```

```python
# aplicar la función a cada paciente
df_pacientes["motivo_consulta"] = df_pacientes.apply(
    lambda fila: asignar_motivo(fila["edad"], fila["ocupacion"]), axis=1
)

df_pacientes.head()
```

```python
# comprobamos los motivos creados
df_pacientes["motivo_consulta"].value_counts()
```

```python
# creamos función para el nivel de urgencia
def asignar_urgencia(motivo):

    if motivo in ["Ansiedad", "Depresión leve", "Duelo"]:
        urgencia = random.choices(niveles_urgencia, weights=[20, 50, 30], k=1)[0]

    elif motivo in ["Estrés académico", "Estrés laboral", "Problemas de pareja"]:
        urgencia = random.choices(niveles_urgencia, weights=[35, 50, 15], k=1)[0]

    else:
        urgencia = random.choices(niveles_urgencia, weights=[50, 40, 10], k=1)[0]

    return urgencia
```

```python
# aplicamos la urgencia a cada paciente 
df_pacientes ["nivel_urgencia"] = df_pacientes["motivo_consulta"].apply(asignar_urgencia)
df_pacientes.head()
```

```python
# comprobamos urgencias 
df_pacientes ["nivel_urgencia"].value_counts()
```

```python
# lo comprobamos en porcentaje 
df_pacientes["nivel_urgencia"].value_counts(normalize= True).round(3)*100
```

```python
# vamos a crear una tabla grande de citas, empezamos por la lita vacia de citas
citas = []
```

```python
# creamos fechas base
fecha_inicio = datetime(2024, 1, 1)
fecha_fin = datetime(2025, 12, 31)
```

```python
# creamos contador de sesiones por paciente 
contador_sesiones = {id_paciente: 0 for id_paciente in ids_pacientes}
```

```python
# generar las 800 citas
for i in range(1, n_citas + 1):

    id_cita = f"C{str(i).zfill(5)}"

    id_paciente = random.choice(ids_pacientes)

    contador_sesiones[id_paciente] += 1
    num_sesion = contador_sesiones[id_paciente]

    datos_paciente = df_pacientes[df_pacientes["id_paciente"] == id_paciente].iloc[0]

    edad = datos_paciente["edad"]
    genero = datos_paciente["genero"]
    ocupacion = datos_paciente["ocupacion"]
    motivo_consulta = datos_paciente["motivo_consulta"]
    nivel_urgencia = datos_paciente["nivel_urgencia"]

    dias_desde_inicio = random.randint(0, (fecha_fin - fecha_inicio).days)
    fecha_solicitud = fecha_inicio + timedelta(days=dias_desde_inicio)

    if nivel_urgencia == "Alto":
        dias_espera = random.randint(1, 7)
    elif nivel_urgencia == "Medio":
        dias_espera = random.randint(4, 15)
    else:
        dias_espera = random.randint(7, 30)

    fecha_cita = fecha_solicitud + timedelta(days=dias_espera)

    if fecha_cita > fecha_fin:
        fecha_cita = fecha_fin

    dia_semana = fecha_cita.strftime("%A")

    traduccion_dias = {
        "Monday": "Lunes",
        "Tuesday": "Martes",
        "Wednesday": "Miércoles",
        "Thursday": "Jueves",
        "Friday": "Viernes",
        "Saturday": "Sábado",
        "Sunday": "Domingo",
    }

    dia_semana = traduccion_dias[dia_semana]

    hora = random.choices(
        [9, 10, 11, 12, 16, 17, 18, 19, 20],
        weights=[8, 10, 10, 8, 12, 16, 16, 12, 8],
        k=1,
    )[0]

    if hora < 14:
        franja_horaria = "Mañana"
    elif hora < 19:
        franja_horaria = "Tarde"
    else:
        franja_horaria = "Noche"

    modalidad = random.choices(modalidades, weights=[45, 55], k=1)[0]

    canal_reserva = random.choices(canales_reserva, weights=[25, 35, 25, 15], k=1)[0]

    if canal_reserva in ["WhatsApp", "Web"]:
        recordatorio_enviado = random.choices(["Sí", "No"], weights=[85, 15], k=1)[0]
    else:
        recordatorio_enviado = random.choices(["Sí", "No"], weights=[60, 40], k=1)[0]

    if recordatorio_enviado == "Sí":
        estado_cita = random.choices(estados_cita, weights=[78, 10, 5, 7], k=1)[0]
    else:
        estado_cita = random.choices(estados_cita, weights=[60, 15, 18, 7], k=1)[0]

    if dia_semana == "Viernes" and franja_horaria in ["Tarde", "Noche"]:
        estado_cita = random.choices(estados_cita, weights=[55, 25, 12, 8], k=1)[0]

    if estado_cita == "No asistida":
        continuidad = random.choices(continuidades, weights=[30, 55, 5, 10], k=1)[0]
    elif estado_cita == "Cancelada":
        continuidad = random.choices(continuidades, weights=[45, 35, 5, 15], k=1)[0]
    elif num_sesion >= 8:
        continuidad = random.choices(continuidades, weights=[45, 15, 35, 5], k=1)[0]
    else:
        continuidad = random.choices(continuidades, weights=[70, 15, 10, 5], k=1)[0]

    if continuidad == "Abandona":
        satisfaccion = random.choices(
            [1, 2, 3, 4, 5], weights=[20, 35, 30, 10, 5], k=1
        )[0]
    elif continuidad == "Alta terapéutica":
        satisfaccion = random.choices([1, 2, 3, 4, 5], weights=[2, 5, 15, 35, 43], k=1)[
            0
        ]
    else:
        satisfaccion = random.choices(
            [1, 2, 3, 4, 5], weights=[5, 10, 25, 35, 25], k=1
        )[0]

    citas.append(
        {
            "id_cita": id_cita,
            "id_paciente": id_paciente,
            "edad": edad,
            "genero": genero,
            "ocupacion": ocupacion,
            "motivo_consulta": motivo_consulta,
            "nivel_urgencia": nivel_urgencia,
            "fecha_solicitud": fecha_solicitud,
            "fecha_cita": fecha_cita,
            "dias_espera": dias_espera,
            "dia_semana": dia_semana,
            "franja_horaria": franja_horaria,
            "modalidad": modalidad,
            "canal_reserva": canal_reserva,
            "recordatorio_enviado": recordatorio_enviado,
            "estado_cita": estado_cita,
            "num_sesion": num_sesion,
            "continuidad": continuidad,
            "satisfaccion": satisfaccion,
        }
    )
```
